```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['warm_encoded'] = label_encoder.fit_transform(df['weather_warm'])
df['cold_encoded'] = label_encoder.fit_transform(df['weather_cold'])
df['dry_encoded'] = label_encoder.fit_transform(df['weather_dry'])
df['wet_encoded'] = label_encoder.fit_transform(df['weather_wet'])
df['cloudy_encoded'] = label_encoder.fit_transform(df['weather_cloudy'])
```

```python
model = svm.SVC(probability = True,  kernel = 'poly' )
```

```python
for season in range (2012, 2019):
  for round in df.loc[df['season'] == season]['round'].unique():
    X = df.loc[(df['season'] == season) & (df['round'] == round)]
    X_train = X[['grid', 'qualifying_time', 'driver_age', 'lap_times', 'warm_encoded', 'cold_encoded', 'dry_encoded', 'wet_encoded', 'cloudy_encoded']]
    Y_train = X.podium

    model.fit(X_train, Y_train)
```
